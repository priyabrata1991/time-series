### There are 8 srong AR lags 0,1,2,3,5,6,13,20 but no seasonal lag

```python
import itertools
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
i=[0,1,2,3,5,6,13,20]
j=[0,1,7]
```

```python
ij=itertools.product(i,j)
```

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
x=[]
l=[]
for i in ij:
    x.append(i)
    try:
        MD=ARMA(df_new2,order=(i))
        result=MD.fit(disp=False)
        l.append(result.aic)
    except:
        l.append(np.nan)
        continue
```

```python
plt.plot(l,marker="+")
plt.show()
```

## Final conclusion

### I will use SARIMAX because there is seasonality in the data

### order will be 6,1,0 I am taking I as 1 beacuse taking first difference makes the data statonary

### seasonal order will be 0,0,1,7 I am taking MA as 1 beacuse strong seasonal lag of 7

```python
from datetime import timedelta
```

```python
my_data=df_new.tail(150)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
test_data=my_data[60:]
```

### I will use rolling window technic for prediction as it gives best predictions

```python
roll_pred=pd.Series()
for i in test_data.index:
    train_data=my_data[:i - timedelta(days=1)]
    model=SARIMAX(train_data, order=(6,1,0), seasonal_order=(0,0,1,7))
    model_fit=model.fit()
    pred=model_fit.predict(i)
    roll_pred[i]=pred.loc[i]
```

```python
plt.figure(figsize=(20,6))
plt.plot(test_data,color='g')
plt.plot(roll_pred,color='b')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(roll_pred-test_data)
plt.axhline(y=0, linestyle="--", color='k')
plt.show()
```

### The resudial error shows no pattern and mean of my resudial error is almost 0

```python
(((abs((test_data-roll_pred)/test_data)).sum())/90)*100
```
